## Glyph Representation

![](images/z.tldraw.png)

```{exercise} Glyph Skeleton

 1. Create a new tldraw document with a unique continuous short curve created with the draw tool (called in the sequel a *glyph*).

 2. Load the document in Python and extract from it the unique record of type `draw`.

 3. Build from this record the arrays `x` and `y` of the points coordinates that describe the curve.

    ⚠️ Make sure that you get the initial values `x[0]` and `y[0]` right!

 5. Plot the curve using an aspect ratio of 1 ; make sure that your conventions for the y axis match those of tldraw.

```

````python
#2/
import json

# Charger le document Tldraw depuis le fichier JSON
tldraw_file_path = 'C:\\Users\\Etudiant\\info-minesparis\\tldraw_new2\\docs\\glyphs.tldr'

with open(tldraw_file_path, 'r', encoding='utf-8') as file:
    tldraw_data = json.load(file)

# Extraire le record de type "draw"
draw_records = [record for record in tldraw_data.get('records', []) if record.get('type') == 'draw']

# Afficher le record de type "draw"
for draw_record in draw_records:
    points = draw_record['props']['segments'][0]['points']
    x_coordinates = [point['x'] for point in points]
    y_coordinates = [point['y'] for point in points]

# Afficher les coordonnées
import matplotlib.pyplot as plt

plt.plot(x_coordinates, y_coordinates)

# Définir un aspect ratio de 1
plt.gca().set_aspect('equal', adjustable='box')
#J'inverse l'axe y car il y a problablement eu une inversion dans les conventions d'axe y 
plt.gca().invert_yaxis()


# Afficher le graphique
plt.show()
````

![](images/z-skeleton-only.jpg)

The curve that you have plotted only represents a "skeleton" of the glyph but not its thickness. 
The thickness itself is computed from an evaluation of the speed at which the pen travels (the faster the pen is, the smaller the curve is ; this is how the pen pressure is simulated).

````python
import matplotlib.pyplot as plt
import perfect_freehand
import numpy as np


freehand_points = np.array(perfect_freehand.get_stroke((x_coordinates, y_coordinates)))

plt.plot(x_coordinates, y_coordinates, 'o', label='Skeleton Points')
plt.plot(freehand_points[:, 0], freehand_points[:, 1], label='Corresponding Contour')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
````
